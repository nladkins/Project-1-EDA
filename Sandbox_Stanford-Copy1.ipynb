```python
# Dependencies and Setup
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import seaborn as sn

# Study data file
stanford_path = "df_us_filtered.csv"

stanford_df = pd.read_csv(stanford_path)
```

```python
# Preview Data Frame
stanford_df.head()
print(len(stanford_df))
```

```python
stanford_df["covid_exercise"].replace("", np.nan, inplace=True)
stanford_df["covid_exercise"].replace("3.0", np.nan, inplace=True)
stanford_df["covid_exercise"].replace("4.0", np.nan, inplace=True)
stanford_df["covid_exercise"].replace("5.0", np.nan, inplace=True)
stanford_df["covid_exercise"].replace("6.0", np.nan, inplace=True)
stanford_df["physical_activities"].replace("", np.nan, inplace=True)
stanford_df["physical_activities"].replace("3.0", np.nan, inplace=True)
stanford_df["physical_activities"].replace("4.0", np.nan, inplace=True)
stanford_df["physical_activities"].replace("5.0", np.nan, inplace=True)
stanford_df["physical_activities"].replace("6.0", np.nan, inplace=True)
```

```python
# # In case there are any empty values, drop surveys that didn't respond to post and pre-covid exercise Q&A.
stanford_df["covid_exercise"].dropna(inplace = True)
stanford_df["physical_activities"].dropna(inplace = True)
print(len(stanford_df))
```

```python
stanford_df = stanford_df.dropna(subset=["physical_activities", "covid_exercise"])
```

```python
print(len(stanford_df))
```

```python
# Indexes
diabetes = stanford_df[stanford_df["diabetes"] == 1].index
cardiovascular_disorders = stanford_df[stanford_df["cardiovascular_disorders"] == 1].index
obesity = stanford_df[stanford_df["obesity"] == 1].index
respiratory_infections = stanford_df[stanford_df["respiratory_infections"] == 1].index
respiratory_disorders = stanford_df[stanford_df["respiratory_disorders_exam"] == 1].index
gastrointenstinal_disorders = stanford_df[stanford_df["gastrointestinal_disorders"] == 1].index
kidney_disease = stanford_df[stanford_df["chronic_kidney_disease"] == 1].index
autoimmune_disease = stanford_df[stanford_df["autoimmune_disease"] == 1].index
fatigue_syndrome = stanford_df[stanford_df["chronic_fatigue_syndrome_a"] == 1].index
```

```python
# yes = stanford_df[stanford_df[physical_activities] == 2]
```

```python
# Generate a pie plot showing the distribution of responses pre-covid

precovid_exercise = stanford_df["physical_activities"].value_counts()

index = [diabetes]

pie_df = pd.DataFrame({"physical activities": precovid_exercise}, index=index)

ax = precovid_exercise.plot.pie(rot=90,  autopct="%1.1f%%", figsize=(5, 5))
plt.title("Physical Activity among Individuals diagnosed with Diabetes (BEFORE COVID)")
plt.axis("equal")
plt.show()
```

```python
# Generate a pie plot showing the distribution of responses pre-covid

duringcovid_exercise = stanford_df["covid_exercise"].value_counts()

index = [diabetes]

pie_df = pd.DataFrame({"pyshical activities": duringcovid_exercise}, index=index)

ax = duringcovid_exercise.plot.pie(rot=0, autopct="%1.1f%%", figsize=(5, 5))
plt.title("Physical Activity among Individuals diagnosed with Diabetes (DURING COVID)")
plt.axis("equal")
plt.show()
```
